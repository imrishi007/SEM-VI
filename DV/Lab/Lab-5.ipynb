# Data Visualization and Interpretation using Seaborn

```python
import seaborn as sn
df = sn.load_dataset("penguins")

df.describe()
```

```python
# 1. Catplot
# o Create a bar catplot to compare average body mass across species and sex.
# o Title: "Average Body Mass by Species and Sex"
# o X-axis: Species
# o Y-axis: Body Mass (g)
# o Hue: Sex

import seaborn as sns
import matplotlib.pyplot as plt

sns.catplot(x="species", y="body_mass_g", hue="sex", data=df, kind="bar",estimator="mean")
plt.title("Average Body Mass by Species and Sex")
plt.show()
```

```python
# Stripplot
# o Visualize flipper length distribution by species and sex.
# o Title: "Flipper Length Distribution by Species and Sex"
# o X-axis: Species
# o Y-axis: Flipper Length (mm)
# o Hue: Sex

sn.stripplot(x='species', y='flipper_length_mm', hue='sex', data=df, jitter=True)
```

```python
# 4. Heatmap
# o Display correlation matrix of numerical features (bill length, bill depth, flipper length, body ,
# mass)
# oAnnotate values, use colorbar to represent correlation strength.
# o Title: "Correlation Heatmap of Penguin Features"

numeric_df = df.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numeric_df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='binary', linewidths=0.5)
plt.title("Correlation Heatmap of Penguin Features")
plt.show()
```

```python
# <!-- Clustermap
# o Create a clustered heatmap to group features based on similarity.
# o Title: "Clustered Correlation Map of Penguin Features" -->

sns.clustermap(correlation_matrix, annot=True, cmap='binary', linewidths=0.5)
plt.title("Clustered Correlation Map of Penguin Features")
plt.show()
```

```python
# 6.Tips Dataset
# Box Plot
# o Compare total bill distribution across days.
# o Title: "Total Bill Distribution by Day"
# o X-axis: Day
# o Y-axis: Total Bill

dataset = sns.load_dataset("tips")
fig,ax = plt.subplots(figsize=(12,10))
sns.boxplot(x='day', y='total_bill', data=dataset,palette='binary',ax=ax)
plt.title("Total Bill Distribution by Day")
plt.show()
```

```python
# Violin Plot
# o Showtip distribution by smoker status.
# o Title: "Tip Distribution by Smoker Status"
# o X-axis: Smoker
# o Y-axis: Tip

sns.violinplot(x='smoker', y='tip', data=dataset, palette='binary')
plt.title("Tip Distribution by Smoker Status")
plt.show()
```

```python
# KDEPlot
# o Plot density of total bill amounts split by gender.
# o Title: "Density of Total Bill by Gender"
# o X-axis: Total Bill
# o Hue: Gender

sns.kdeplot(data=dataset, x='total_bill', hue='sex', fill=True, palette='binary')
plt.title("Density of Total Bill by Gender")
plt.show()
```

```python
# Joint Plot
# o Explore relationship between total bill and tip.
# o Title: "Total Bill vs Tip Relationship"
# o X-axis: Total Bill
# o Y-axis: Tip
# o Hue: Gender

sns.jointplot(data=dataset, x='total_bill', y='tip', hue='sex', palette='binary')
plt.suptitle("Total Bill vs Tip Relationship", y=1.02)
plt.show() 
```

```python
# IRIS Dataset
# Pairplot
# o Create pairwise relationships among all numeric features.
# o Title: "Pairwise Relationships in Iris Dataset"
# o Hue: Species

iris_df = sns.load_dataset("iris")

sns.pairplot(iris_df, hue='species', palette='binary')
plt.suptitle("Pairwise Relationships in Iris Dataset", y=1.02)
plt.show()
```

```python
# SwarmPlot
# o Visualize petal length distribution across species.
# o Title: "Petal Length Distribution by Species"
# o X-axis: Species
# o Y-axis: Petal Length

sns.swarmplot(data=iris_df, x="species", y="petal_length", orient="v")
```

```python
# Bubble Chart (Scatterplot with size)
# o Sepal length vs sepal width
# o bubble size = petal length
# o hue = species.
# o Title: "Sepal vs Petal Dimensions in Iris"

sns.scatterplot(data=iris_df, x='sepal_length', y='sepal_width', size='petal_length', hue='species', sizes=(20, 200), palette='binary', alpha=0.7)
plt.title("Sepal vs Petal Dimensions in Iris")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```
