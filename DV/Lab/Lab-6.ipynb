Lab 6-Data Visualization and Interpretation using Plotly 

```python
import plotly.express as px
import pandas as pd
```

```python
data = px.data.gapminder()
```

```python
# 1. Scatter Plot with Slider (Animation) 
# o Plot GDP per capita (gdpPercap) vs life expectancy (lifeExp) for all countries. 
# o Use population (pop) to size the points. 
# o Color points by continent. 
# o Add a slider to move through different years. 
# o Title: "GDP vs Life Expectancy Over Time". 
# o Label the x-axis as "GDP per Capita (USD)" and the y-axis as "Life Expectancy (Years)". 
# o Ensure hover tooltips display country names. 

fig = px.scatter(data, x="gdpPercap", y="lifeExp",size="pop",color="continent", animation_frame="year",title="GDP vs Life Expectancy Over Time", labels={"gdpPercap": "GDP per Capita (USD)", "lifeExp": "Life Expectancy (Years)"}, hover_name="country")
fig.show()
```

```python
# 5. Animated Scatter Plot with Play/Pause Controls 
# o Create an animated scatter plot of GDP per capita vs life expectancy across years. 
# o Use population (pop) for point size. 
# o Color points by continent. 
# o Add play/pause controls to animate automatically. 
# o Title: "Global Development Trends (19522007)". 
# o Label axes appropriately. 

fig2 = px.scatter(data, x="gdpPercap", y="lifeExp", size="pop", color="continent", animation_frame="year", title="Global Development Trends (19522007)", labels={"gdpPercap": "GDP per Capita (USD)", "lifeExp": "Life Expectancy (Years)"}, hover_name="country")
fig2.update_layout(updatemenus=[dict(type="buttons", showactive=False, buttons=[dict(label="Play", method="animate", args=[None, {"frame": {"duration": 500, "redraw": True}, "fromcurrent": True}]), dict(label="Pause", method="animate", args=[[None], {"frame": {"duration": 0, "redraw": False}, "mode": "immediate", "transition": {"duration": 0}}])])])
fig2.show()
```

```python
# 6. 3D Scatter Plot 
# o Plot GDP per capita (gdpPercap), life expectancy (lifeExp), and population (pop) in 2007. 
# o Use continent for color coding. 
# o Size points by population. 
# o Title: "3D Scatter: GDP, Life Expectancy, Population (2007)". 
# o Label axes appropriately.

df_2007 = data[data.year==2007]
c = px.scatter_3d(df_2007, x="gdpPercap", y="lifeExp", z="pop", color="country", size="pop", title="3D Scatter: GDP, Life Expectancy, Population (2007)", labels={"gdpPercap": "GDP per Capita (USD)", "lifeExp": "Life Expectancy (Years)", "pop": "Population"})
c.show()
```

```python
# 7. Heatmap 
# o Compute the correlation matrix for lifeExp, gdpPercap, and pop. 
# o Plot the matrix as a heatmap. 
# o Add numeric values inside each cell. 
# o Title: "Correlation Heatmap of Development Indicators". 
# o Use a diverging color scale for better contrast. 

corr_matrix = data[["lifeExp", "gdpPercap", "pop"]].corr()
heatmap = px.imshow(corr_matrix, text_auto=True, color_continuous_scale="RdBu", title="Correlation Heatmap of Development Indicators")
heatmap.show()
```

```python
# 8. Pie Chart 
# o Create a pie chart showing population share (pop) by continent in 2007. 
# o Add percentage labels to each slice. 
# o Title: "Population Share by Continent (2007)". 
# o Use contrasting colors for slices. 

df_2007_continent = df_2007.groupby("continent")["pop"].sum().reset_index()
pie_chart = px.pie(df_2007_continent, values="pop", names="continent", 
                   title="Population Share by Continent (2007)", color_discrete_sequence=px.colors.qualitative.Set3)
pie_chart.update_traces(textposition="inside", textinfo="percent+label")
pie_chart.show()
```

```python
# 9. Treemap of Population 
# o Title: "Population Treemap (2007)" 
# o Size blocks by population. 
# o Color blocks by continent. 
# o Hover tooltips should show country names.

treemap = px.treemap(df_2007, path=["continent", "country"], values="pop", color="continent", title="Population Treemap (2007)", color_discrete_sequence=px.colors.qualitative.Set3)
treemap.show()
```
