# Lab-3 Data Visualization and Interpretation Using Matplotlib-II

### 1. Write code for line chart to generate 70 numbers with X- axis and Y-axis is mx+c, were m=2 and c=0.3,
### Draw the four different line charts using matplotlib library. (Take Y1=X, Y2=X**2, Y3=X**3 and
### Y4=sqrt(X)).

```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Generate 70 numbers for X-axis
X = np.linspace(0, 10, 70)
m , c = 2, 0.3
Y = m * X + c
Y1 = X
Y2 = X**2
Y3 = X**3
Y4 = np.sqrt(X)

# Create subplots
fig, ax = plt.subplots(2,2, figsize=(10,8))

# Plot Y1 = X
ax[0,0].plot(X,Y1, color='blue',label= 'Y1 = X')
ax[0,0].set_title('Line Chart of Y1 = X')
ax[0,0].set_xlabel('X-axis')
ax[0,0].set_ylabel('Y1-axis')
ax[0,0].legend()
ax[0,0].grid()

# Plot Y2 = X^2
ax[0,1].plot(X,Y2, color='green',label= 'Y2 = X^2')
ax[0,1].set_title('Line Chart of Y2 = X^2')
ax[0,1].set_xlabel('X-axis')
ax[0,1].set_ylabel('Y2-axis')
ax[0,1].legend()
ax[0,1].grid()

# Plot Y3 = X^3
ax[1,0].plot(X,Y3, color='red',label= 'Y3 = X^3')
ax[1,0].set_title('Line Chart of Y3 = X^3')
ax[1,0].set_xlabel('X-axis')
ax[1,0].set_ylabel('Y3-axis')
ax[1,0].legend()
ax[1,0].grid()

# Plot Y4 = sqrt(X)
ax[1,1].plot(X,Y4, color='purple',label= 'Y4 = sqrt(X)')
ax[1,1].set_title('Line Chart of Y4 = sqrt(X)')
ax[1,1].set_xlabel('X-axis')
ax[1,1].set_ylabel('Y4-axis')
ax[1,1].legend()
ax[1,1].grid()

plt.tight_layout()
plt.show()
```

```python
# 2. Plot the function y = x- 4x + 3 for x ranging from-2 to 6. Identify and annotate the minimum point
# of the graph. Add appropriate labels, title, and grid

X = np.linspace(-2, 6, 100)
Y = X**2 - 4*X + 3

#X_Min = -b/2a -> -(-4)/2 == 2
min_x = 2
min_y = min_x**2 - 4*min_x + 3 # -1
plt.figure(figsize=(8,6))
plt.plot(X, Y, label='y = x - 4x + 3', color='orange')
plt.scatter(min_x, min_y, color='red')  
plt.title('Plot of y = x - 4x + 3')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()
plt.legend()
plt.show()
```

```python
# 3. Plot sin(x) and cos(x) from 0 to 2 on the same graph. Use fill_between() to shade the area between the two curves where sin(x) is greater than cos(x). Add labels and a title.
X = np.linspace(0,2*math.pi, 120)
Y1 = np.sin(X)
Y2 = np.cos(X)
plt.figure(figsize=(10,6))
plt.plot(X, Y1, label='sin(x)', color='blue')
plt.plot(X, Y2, label='cos(x)', color='green')
plt.fill_between(X,Y1,Y2, where=(Y1>Y2) , color='lightblue', label='Area where sin(x) > cos(x)')
plt.title('Plot of sinx and cosx')
plt.xlabel('X-axis (radians)')
plt.ylabel('Y-axis')
plt.legend()
plt.grid()
plt.show()
```

```python
#4. Consider monthly temperature data for the year 2023 and plot it using a time-series line graph.
# months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
# temperature = [12, 14, 18, 24, 30, 34, 32, 31, 28, 22, 16, 13]

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
temperature = [12, 14, 18, 24, 30, 34, 32, 31, 28, 22, 16, 13]
plt.figure(figsize=(10,6))
plt.plot(months, temperature, marker='x', color='blue', label='Temperature (2023)')
plt.title('Monthly Temperature Data for 2023')
plt.xlabel('Months')
plt.ylabel('Temperature (C)')
plt.grid()
plt.legend()
plt.show()
```

```python
# Plot three mathematical functions on the same graph:
# y=x(Red,DashedLine)
# y=2x+1(Blue,Solid Line)
# y=sqrt(x) (Green, Dotted Line)

X = np.linspace(0,10,100)
Y1,Y2,Y3 = X**2, 2*X + 1 , np.sqrt(X)
plt.figure(figsize=(10,6))
plt.plot(X, Y1, color='red', linestyle='--', label='y = x')
plt.plot(X, Y2, color='blue', linestyle='-', label='y = 2x + 1')
plt.plot(X, Y3, color='green', linestyle=':', label='y = sqrt(x)')
plt.title('Plot of Three Mathematical Functions')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid()
plt.show()
```

```python
# 6. Create a dual-axis plot where:
# The primary y-axis represents the monthly rainfall (in mm) in 2023 as a bar chart.
# The secondary y-axis represents the average temperature (in C) as a line graph on the same plot.
# months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
# rainfall = [78, 62, 55, 41, 32, 140, 300, 280, 150, 90, 85, 80]
# temperature = [12, 14, 18, 24, 30, 34, 32, 31, 28, 22, 16, 13]
# Label both y-axes and use different colors for bars and the line.

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
rainfall = [78, 62, 55, 41, 32, 140, 300, 280, 150, 90, 85, 80]
temperature = [12, 14, 18, 24, 30, 34, 32, 31, 28, 22, 16, 13]

fig, ax1 = plt.subplots(figsize=(10,6))
ax1.bar(months,rainfall, color='skyblue',label='Rainfall in mm')
ax1.set_xlabel('Months')
ax1.set_ylabel('Rainfall (mm)', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.set_title('Monthly Rainfall and Average Temperature in 2023')
ax1.grid()

ax2 = ax1.twinx()
ax2.plot(months,temperature,color = 'red', label='Average Temperature (C)')
ax2.set_ylabel('Temperature (C)', color='red')
ax2.tick_params(axis='y', labelcolor='red')
fig.tight_layout()
plt.show()
```
