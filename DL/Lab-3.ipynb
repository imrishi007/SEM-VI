## Import Required Libraries

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
```

## Program 1: Tensor Creation and Basic Operations

**Aim:** To understand tensor creation and perform basic arithmetic operations using TensorFlow.

```python
# Scalar
a = tf.constant(5)
b = tf.constant(3)

# Vector
v1 = tf.constant([1, 2, 3])
v2 = tf.constant([4, 5, 6])

# Matrix
m1 = tf.constant([[1, 2], [3, 4]])
m2 = tf.constant([[5, 6], [7, 8]])

print("Scalar Addition:", a + b)
print("Vector Addition:", v1 + v2)
print("Matrix Multiplication:\n", tf.matmul(m1, m2))
```

## Program 2: Linear Regression Using TensorFlow

**Aim:** To implement a simple linear regression model using TensorFlow.

```python
# Data
x = np.array([1, 2, 3, 4, 5], dtype=float)
y = np.array([2, 4, 6, 8, 10], dtype=float)

# Model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=1, input_shape=[1])
])

model.compile(optimizer='sgd', loss='mean_squared_error')
model.fit(x, y, epochs=100, verbose=0)

# Prediction
print("Prediction for x=6:", model.predict(np.array([6.0])))
```

## Program 3: Simple Neural Network for Classification

**Aim:** To build and train a neural network for binary classification.

```python
# Dataset
X, y = make_classification(n_samples=500, n_features=2, n_classes=2, n_redundant=0, random_state=42)

# Model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(8, activation='relu', input_shape=(2,)),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(X, y, epochs=20, verbose=0)
loss, acc = model.evaluate(X, y, verbose=0)
print("Accuracy:", acc)
```

Modify Program 1 to perform division and transpose operations.

```python
# Scalar division
a = tf.constant(10)
b = tf.constant(2)
print("Scalar Division:", a / b)

# Vector division
v1 = tf.constant([10, 20, 30], dtype=tf.float32)
v2 = tf.constant([2, 4, 5], dtype=tf.float32)
print("Vector Division:", v1 / v2)

# Matrix transpose
m1 = tf.constant([[1, 2], [3, 4]])
print("Original Matrix:\n", m1)
print("Transposed Matrix:\n", tf.transpose(m1))

# Matrix division
m2 = tf.constant([[10, 20], [30, 40]], dtype=tf.float32)
m3 = tf.constant([[2, 4], [5, 8]], dtype=tf.float32)
print("Matrix Division:\n", m2 / m3)
```

## Exercise 2: Modified Linear Regression with Different Dataset

Change the dataset in Program 2 and observe loss variation.

```python
# Modified dataset with different linear relationship
x = np.array([1, 2, 3, 4, 5, 6, 7, 8], dtype=float)
y = np.array([3, 7, 11, 15, 19, 23, 27, 31], dtype=float)

# Model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=1, input_shape=[1])
])

model.compile(optimizer='sgd', loss='mean_squared_error')

# Train and store history
history = model.fit(x, y, epochs=100, verbose=0)

# Plot loss variation
plt.figure(figsize=(10, 5))
plt.plot(history.history['loss'])
plt.title('Model Loss During Training')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.grid(True)
plt.show()

# Prediction
print("Prediction for x=10:", model.predict(np.array([10])))
print("Final Loss:", history.history['loss'][-1])
```

## Exercise 3: Neural Network with Additional Hidden Layer

Add an extra hidden layer in Program 3 and compare accuracy.

```python
# Dataset
X, y = make_classification(n_samples=500, n_features=2, n_classes=2, n_redundant=0, random_state=42)

# Original model
model_original = tf.keras.Sequential([
    tf.keras.layers.Dense(8, activation='relu', input_shape=(2,)),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model_original.compile(optimizer='adam',
                       loss='binary_crossentropy',
                       metrics=['accuracy'])

model_original.fit(X, y, epochs=20, verbose=0)
loss_orig, acc_orig = model_original.evaluate(X, y, verbose=0)

# Modified model with additional hidden layer
model_modified = tf.keras.Sequential([
    tf.keras.layers.Dense(8, activation='relu', input_shape=(2,)),
    tf.keras.layers.Dense(4, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model_modified.compile(optimizer='adam',
                       loss='binary_crossentropy',
                       metrics=['accuracy'])

model_modified.fit(X, y, epochs=20, verbose=0)
loss_mod, acc_mod = model_modified.evaluate(X, y, verbose=0)

print("Original Model Accuracy:", acc_orig)
print("Modified Model Accuracy:", acc_mod)
print("Accuracy Difference:", acc_mod - acc_orig)
```

## Exercise 4: Optimizer Comparison (Adam vs SGD)

Replace optimizer adam with sgd and analyze performance.

```python
import time

# Dataset
X, y = make_classification(n_samples=500, n_features=2, n_classes=2, n_redundant=0, random_state=42)

# Model with Adam optimizer
model_adam = tf.keras.Sequential([
    tf.keras.layers.Dense(8, activation='relu', input_shape=(2,)),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model_adam.compile(optimizer='adam',
                   loss='binary_crossentropy',
                   metrics=['accuracy'])

start_time = time.time()
history_adam = model_adam.fit(X, y, epochs=20, verbose=0)
adam_time = time.time() - start_time
loss_adam, acc_adam = model_adam.evaluate(X, y, verbose=0)

# Model with SGD optimizer
model_sgd = tf.keras.Sequential([
    tf.keras.layers.Dense(8, activation='relu', input_shape=(2,)),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model_sgd.compile(optimizer='sgd',
                  loss='binary_crossentropy',
                  metrics=['accuracy'])

start_time = time.time()
history_sgd = model_sgd.fit(X, y, epochs=20, verbose=0)
sgd_time = time.time() - start_time
loss_sgd, acc_sgd = model_sgd.evaluate(X, y, verbose=0)

# Performance comparison
print("Adam Optimizer:")
print("  Accuracy:", acc_adam)
print("  Training Time:", adam_time, "seconds")

print("\nSGD Optimizer:")
print("  Accuracy:", acc_sgd)
print("  Training Time:", sgd_time, "seconds")

print("\nComparison:")
print("  Accuracy Difference:", acc_adam - acc_sgd)
print("  Time Difference:", sgd_time - adam_time, "seconds")
```

## Exercise 5: Multi-Class Classification with Softmax

Implement multi-class classification using Softmax activation.

```python
# Dataset with 3 classes
X, y = make_classification(n_samples=600, n_features=4, n_classes=3, 
                           n_informative=3, n_redundant=0, random_state=42)

# Convert labels to categorical
y_categorical = tf.keras.utils.to_categorical(y, num_classes=3)

# Model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(16, activation='relu', input_shape=(4,)),
    tf.keras.layers.Dense(8, activation='relu'),
    tf.keras.layers.Dense(3, activation='softmax')
])

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

history = model.fit(X, y_categorical, epochs=30, verbose=0)
loss, acc = model.evaluate(X, y_categorical, verbose=0)

print("Multi-Class Classification Accuracy:", acc)
print("Final Loss:", loss)

# Sample prediction
sample = X[:5]
predictions = model.predict(sample)
print("\nSample Predictions:")
for i, pred in enumerate(predictions):
    print(f"Sample {i+1}: Class {np.argmax(pred)} (Probabilities: {pred})")
```

## Exercise Questions

1. **What is a tensor?**  
   A tensor is a multi-dimensional array of numerical values that serves as the fundamental data structure in TensorFlow. It can represent scalars (0D), vectors (1D), matrices (2D), or higher-dimensional data, and is used for computations in neural networks.

2. **Difference between NumPy arrays and TensorFlow tensors:**  
   NumPy arrays are fixed-size, CPU-based arrays designed for general numerical computing and scientific calculations. TensorFlow tensors are similar multi-dimensional arrays but are part of TensorFlow's computational graph, support dynamic shapes, GPU acceleration, and automatic differentiation for machine learning tasks.

3. **What is an optimizer?**  
   An optimizer is an algorithm that adjusts the parameters (weights and biases) of a neural network during training to minimize the loss function. It uses techniques like gradient descent (e.g., SGD, Adam) to iteratively update parameters based on the computed gradients.

4. **Explain loss functions:**  
   Loss functions measure the difference between the model's predictions and the actual target values, quantifying how well the model performs. They guide training by providing a value to minimize. Examples include mean squared error (MSE) for regression tasks and binary/cross-entropy for classification tasks.

5. **What is backpropagation?**  
   Backpropagation is an algorithm used in neural networks to compute the gradients of the loss function with respect to each parameter (weight and bias) in the network. It works backward from the output layer to the input layer, enabling the optimizer to update parameters and improve the model's accuracy.
